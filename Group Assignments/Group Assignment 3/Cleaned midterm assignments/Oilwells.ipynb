# Oil Wells in LA County
This notebook describes our initial process to map the oil wells in LA county, prior to adding our demographic data for analysis.

I started by deleting some of the columns we will not be using in excel, as we had issues initially with the size of the dataset messing up the ability to do any analysis.

```python
# added pandas library
import pandas as pd
import geopandas as gpd
```

```python
# Had pandas read the csv file of oil wells
df = pd.read_csv('Data/Oil_Wells.csv')
```

# Active wells

Because the dataset was so large, we had issues with trying analyses initially. We decided to narrow to active wells because this is a much smaller amount of data but useful filter for understanding oil wells in the city.

```python
#Here is a chart we created to show that there are a significant amount of plugged wells within the dataset.
import matplotlib.pyplot as plt
import seaborn as sns

# Filter the dataframe for only active ('A') and plugged ('P') wells
filtered_df = df[df['WellStatus'].isin(['A', 'P'])]

# Replace 'A' with 'Active' and 'P' with 'Plugged' in the 'WellStatus' column
filtered_df.loc[:, 'WellStatus'] = filtered_df['WellStatus'].replace({'P': 'Plugged', 'A': 'Active'})

# Plot the count of active vs. plugged wells
plt.figure(figsize=(8, 6))
sns.countplot(data=filtered_df, x='WellStatus')
plt.title('Count of Active vs. Plugged Wells')
plt.show()
```

I then created a geodataframe for the information.

```python
gdf = gpd.GeoDataFrame(df, 
                       crs='epsg:4326',
                       geometry=gpd.points_from_xy(df.Longitude, df.Latitude))
```

```python
gdf_act = gdf.drop(gdf[gdf['WellStatus'] != 'A'].index)
```

```python
gdf_act.info()
```

I also confirmed that the gdf of active wells was in the mercator projection.

```python
gdf_act.crs
```

# Mapping

Next, I made a basic maps using plotly express to show the active wells in LA with a background showing roads and other features, including greenspace, near these wells.

```python
import plotly.express as px
```

```python
fig = px.scatter_mapbox(gdf_act,
                        lat='Latitude',
                        lon='Longitude',
                        color_discrete_sequence=["Black"],
                        mapbox_style="open-street-map")

# options on the layout
fig.update_layout(
    width = 800,
    height = 800,
    title = "Interactive Map of Oil Wells in LA County",
    title_x=0.5 # aligns title to center
)

fig.show()
```
