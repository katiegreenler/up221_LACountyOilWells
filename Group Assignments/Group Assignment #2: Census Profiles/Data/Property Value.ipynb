```python
#Importing the geopandas package to conduct GIS analsis
import geopandas as gpd
```

```python
#Read the existing 2020 Geojson file
tracts=gpd.read_file('Census_Tracts_2020.geojson')
tracts.head()
```

```python
#Plot the Census 2020 Census Tract Geojson Dataframe 
tracts.plot(figsize=(12,10))
```

```python
#More information about the 2020 Census Geojson Files
tracts.info(verbose=True, show_counts=True)
```

```python
#Limting the columns to only CT20 and Geometry
tracts = tracts[['CT20','geometry']]
tracts.head()
```

```python
#creating a FISP column for the data we will merge
tracts['FIPS'] ='06' + '037' + tracts['CT20']
```

```python
#sampling the Tract column
tracts.head()
```

```python
#Merged the 2020 Census Gejston File and 2022 Property Value File 
tracts_race=tracts.merge(df2,on="FIPS")
```

```python
#Sampling the merged data
tracts_race.head()
```

```python
#Mapping the % porportion of the All Owner-Occupied Units data, used equal_interval since the Owner-Occupeid Units Data is already %
tracts_race.plot(figsize=(12,10),
                 column='All Owner-Occupied Units',
                 legend=True, 
                 scheme='equal_interval')
```

```python
#Mapping the % porportion of the Owner-Occupied Units Valued at $1 Million, used equal_interval since the Owner-Occupeid Units Data is already %
tracts_race.plot(figsize=(12,10),
                 column='More than $1Million',
                 legend=True, 
                 scheme='equal_interval')
```

```python
#Mapping the % porportion of the Owner-Occupied Units Valued at less than $20K, used equal_interval since the Owner-Occupeid Units Data is already %
tracts_race.plot(figsize=(12,10),
                 column='Less than $20K',
                 legend=True, 
                 scheme='equal_interval')
```

```python
#Mapping the % porportion of the Owner-Occupied Units Valued between $300K-499K, used equal_interval since the Owner-Occupeid Units Data is already %
tracts_race.plot(figsize=(12,10),
                 column='300K-499K',
                 legend=True, 
                 scheme='equal_interval')
```
